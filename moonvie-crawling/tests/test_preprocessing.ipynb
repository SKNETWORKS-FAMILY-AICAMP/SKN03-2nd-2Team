```python
df = pd.read_csv('WeeklyBoxOfficeData.csv')
```

```python
df.info()
```

```python
# 중복 행 제거
df = df.drop_duplicates()
df.info()
```

```python
# 개봉일 column이 빈칸인 rows 찾아 제거
blank_value = ' '
column_name = '개봉일'
cnt = 0
if blank_value in df[column_name].values:
    matching_rows_idx = df[df[column_name] == blank_value].index
    for _ in matching_rows_idx:
        cnt += 1
    print(f'blank count: {cnt}\n {matching_rows_idx}')
    print('*'*30)

for matching_row in matching_rows_idx:
    df = df.drop(matching_row)

df.info()
```

```python
# 2017년 이전 개봉작 제거
df['개봉일'] = pd.to_datetime(df['개봉일'])
df = df[df['개봉일'] >= '2017-01-01']
df.info()
```

```python
# '상영기간'에서 시작일과 종료일을 분리
df['상영시작일'] = df['상영기간'].apply(lambda x: x.split('~')[0])
df['상영종료일'] = df['상영기간'].apply(lambda x: x.split('~')[1])

# 시작일과 종료일을 datetime 형식으로 변환
df['상영시작일'] = pd.to_datetime(df['상영시작일'], format='%Y%m%d')
df['상영종료일'] = pd.to_datetime(df['상영종료일'], format='%Y%m%d')

# 불필요한 column 제거
df.drop(columns=['상영기간','순번', '박스오피스 순위', '박스오피스종류', '신규진입여부'], inplace=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.sort_values(by='조회연도및주차', ascending=False)
```

```python
df.info()
```

```python
from datetime import datetime

# 상영시작일과 상영종료일 사이의 모든 월을 추출
def get_months_range(start_date, end_date):
    months = set()
    current_date = start_date

    while current_date <= end_date:
        months.add(current_date.strftime('%Y-%m'))
        if current_date.month == 12:
            current_date = datetime(current_date.year + 1, 1, 1)
        else:
            current_date = datetime(current_date.year, current_date.month + 1, 1)

    return sorted(list(months))

# 빈 리스트를 만들어 새로운 행을 저장할 준비
new_df = []

# 각 행을 반복하며 상영 기간을 월별로 분리
for index, row in df.iterrows():
    months = get_months_range(row['상영시작일'], row['상영종료일'])
    for month in months:
        new_row = row.copy()
        new_row['상영월'] = month
        new_df.append(new_row)

# 새로운 DataFrame 생성
new_df = pd.DataFrame(new_df)
new_df
```

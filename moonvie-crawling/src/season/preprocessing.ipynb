```python
import pandas as pd
import os

df = pd.read_csv('C:/dev/github/SKN03-2nd-2Team/moonvie-crawling/data/season/bf_preprocessing/전국_WeeklyBoxOfficeData.csv')
```

```python
df['지역'] = '전국'
df.info()
```

```python
df.info()
```

```python
# 중복 행 제거
df = df.drop_duplicates()
df.info()
```

```python
# 개봉일 column이 빈칸인 rows 찾아 제거
blank_value = ' '
column_name = '개봉일'
cnt = 0
if blank_value in df[column_name].values:
    matching_rows_idx = df[df[column_name] == blank_value].index
    for _ in matching_rows_idx:
        cnt += 1
    print(f'blank count: {cnt}\n {matching_rows_idx}')
    print('*'*30)

for matching_row in matching_rows_idx:
    df = df.drop(matching_row)

df.info()
```

```python
# 2017년 이전 개봉작 제거
df['개봉일'] = pd.to_datetime(df['개봉일'])
df = df[df['개봉일'] >= '2017-01-01']
df.info()
```

```python
df.head()
```

```python
# '영화대표코드'를 기준으로 그룹화한 후, '누적관객수'가 가장 높은 행의 인덱스를 선택
rows = df.groupby('영화대표코드')['누적 관객수'].idxmax()

# 선택된 인덱스의 행들만 남기기
df = df.loc[rows]

# 인덱스 재설정 (옵션)
df = df.reset_index(drop=True)
df
```

```python
from datetime import datetime

# '상영시작일'과 '상영종료일' 컬럼 생성
df[['상영시작일', '상영종료일']] = df['상영기간'].str.split('~', expand=True)

# '상영시작일'과 '상영종료일' 컬럼을 날짜 형식으로 변환
df['상영시작일'] = pd.to_datetime(df['상영시작일'], format='%Y%m%d')
df['상영종료일'] = pd.to_datetime(df['상영종료일'], format='%Y%m%d')

# 상영 기간을 월별로 분리하는 함수 정의
def get_months_range(start_date, end_date):
    months = set()
    current_date = start_date

    while current_date <= end_date:
        months.add(current_date.month)
        if current_date.month == 12:
            current_date = datetime(current_date.year + 1, 1, 1)
        else:
            current_date = datetime(current_date.year, current_date.month + 1, 1)

    return sorted(list(months))

# '상영연도'와 '상영월' 컬럼 생성
df['상영연도'] = df['상영시작일'].dt.year
df['상영월'] = df.apply(lambda row: ', '.join(map(str, get_months_range(row['상영시작일'], row['상영종료일']))), axis=1)

df = df[['영화대표코드', '제목', '개봉일', '해당일자 상영작 대비 해당영화 매출비율',
         '누적 매출액', '누적 관객수', '상영시작일', '상영종료일', '상영연도', '상영월', '지역']]
```

# csv 통합

```python
import pandas as pd
import os

DATA_PATH = 'C:/dev/github/SKN03-2nd-2Team/moonvie-crawling/data/season/af_preprocessing'
folder = os.listdir(DATA_PATH)
os.chdir(DATA_PATH)

# 빈 리스트 생성 - 각 CSV 파일의 데이터프레임을 저장
dataframes = []

# 디렉토리 내 모든 CSV 파일 읽기
for file in folder:
    if file.endswith('.csv'):  # 파일이 .csv로 끝나는지 확인
        df = pd.read_csv(file, encoding='utf-8-sig')
        dataframes.append(df)  # 리스트에 데이터프레임 추가

# 모든 데이터프레임을 하나로 병합
merged_df = pd.concat(dataframes, ignore_index=True)

# 병합된 데이터프레임을 새로운 CSV 파일로 저장
merged_df.to_csv('Total_WeeklyBoxOfficeData.csv', index=False, encoding='utf-8-sig')

print('모든 CSV 파일이 성공적으로 병합되었습니다.')
```
